### Linear Equation Construction
The first example will demonstrate calculation of the slope for a linear equation using Example 5 from Section 1.1 of Lial.

```python
x1= 5.0
y1= 4.0
x2= -10.0
y2= -2.0
slope= (y2-y1)/(x2-x1)
slope
```

The next example will show how to use the calculated slope to form the linear equation and calculate a result.  

```python
x= 1.0
y= y1 + slope*(x-x1)
print('\nValue of y if x is 1.0 equals {}'.format(y))
```

The next example will show how to solve Example 6 Section 1.2 of Lial graphically using Python.  Note that integers and floating point numbers can combined in the same equation.  The result is a floating point number.  The figure() statement separates the following plot from the previous plot. Two windows will appear with separate plots one behind the other.

```python
x= linspace(0,50,100)
y= 20*x+100.0
z= 24*x
```
